# Sales Report

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename = "sales_january.xlsx"
```

```python
data = pd.read_excel(filename, index_col=0)
data
```

```python
plt.style.use('seaborn-darkgrid')
ax = data.plot(kind='bar', label='index', figsize=(15, 6), color='crimson', rot=0, fontsize=15)
ax.set_title(f'Sales report from {filename}', fontsize=25, fontweight='bold')
ax.set_xlabel('Employees', fontsize=15)
ax.set_ylabel('Sales in $', fontsize=15);
```
